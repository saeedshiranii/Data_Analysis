# Input and Output Variables
Forgetting to assign the result of a function call to a variable can be very annoying.
Fortunately, IPython stores references to both the input (the text that you type) and
output (the object that is returned) in special variables. The previous two outputs are
stored in the _ (one underscore) and __ (two underscores) variables, respectively:

In [556]: 2 ** 27
Out[556]: 134217728
In [557]: _
Out[557]: 134217728
Input variables are stored in variables named like _iX, where X is the input line number.
For each such input variables there is a corresponding output variable _X. So after input
line 27, say, there will be two new variables _27 (for the output) and _i27 for the input.
In [26]: foo = 'bar'
In [27]: foo
Out[27]: 'bar'
In [28]: _i27
Out[28]: u'foo'
In [29]: _27
Out[29]: 'bar'
Since the input variables are strings, that can be executed again using the Python
exec keyword:
In [30]: exec _i27
Several magic functions allow you to work with the input and output history. %hist is
capable of printing all or part of the input history, with or without line numbers.
%reset is for clearing the interactive namespace and optionally the input and output
caches. The %xdel magic function is intended for removing all references to a particu-
lar object from the IPython machinery. See the documentation for both of these magics
for more details.
When working with very large data sets, keep in mind that IPython’s
input and output history causes any object referenced there to not be
garbage collected (freeing up the memory), even if you delete the vari-
ables from the interactive namespace using the del keyword. In such
cases, careful usage of %xdel and %reset can help you avoid running into
memory problems.

# Logging the Input and Output
IPython provides a feature that allows you to log your entire console session, including both the input commands you type and the corresponding output generated by those commands. This feature can be helpful for keeping a record of your work, documenting your analysis, or sharing your session with others. The process of logging the input and output is initiated using the `%logstart` magic command.

Here's how you can use it:

1. **Activating Logging**: To start logging your IPython session, you can type the following command:

   ```
   %logstart
   ```

   When you run this command, IPython will activate auto-logging, which means it will record both the current session's state and any future input you provide.

2. **Logging Configuration**:
   After you run `%logstart`, IPython will provide you with information about the logging configuration it's using. This includes details such as:
   
   - `Filename`: The name of the file where the log will be saved (`ipython_log.py` in this case).
   - `Mode`: The mode in which the log file is managed (e.g., rotate, append).
   - `Output logging`: Whether output logging (result of executed commands) is enabled.
   - `Raw input log`: Whether raw input (including whitespace and comments) is logged.
   - `Timestamping`: Whether timestamps are added to each logged command.
   - `State`: The current state of logging (active or inactive).

3. **Enabling and Disabling Logging**: You can enable or disable IPython logging at any point during your session. If you decide you want to save your entire session from a certain point onward, you can use `%logstart`. If you want to stop logging, you can use `%logoff` or `%logstop`. If you later want to resume logging, you can use `%logon`.

4. **Advanced Options**: The `%logstart` command has additional options that allow you to customize the logging behavior. For example, you can specify the filename for the log, change the log mode, enable output logging, include timestamps, and more.

   You can find more information about these options in the docstring for `%logstart`. You can access the docstring by typing `?%logstart`.

5. **Companion Functions**: Along with `%logstart`, there are other related magic commands you can use to manage logging, such as `%logoff`, `%logon`, `%logstate`, and `%logstop`. These commands provide control over the logging state and options.

Remember that logging is a useful feature, but it can consume storage space if not managed properly. Use it wisely to document important sessions or experiments.

As with any IPython magic command, you can explore the detailed documentation by typing `?%command_name` to get more information about its usage and options.

# Directory Bookmark System



**IPython's Directory Bookmarking System**

IPython provides a convenient directory bookmarking system that allows you to save aliases for frequently used directories. These aliases, known as bookmarks, enable you to easily navigate and jump between different directories without having to remember or type out the full paths.

**Creating Bookmarks:**

For instance, if you frequently use Dropbox and want to quickly switch to that directory, you can create a bookmark like this:

```python
In [1]: %bookmark db /home/wesm/Dropbox/
```

This assigns the alias `db` to the directory `/home/wesm/Dropbox/`.

**Using Bookmarks:**

Once you've defined bookmarks, you can utilize them with the `%cd` magic command. For instance, using the bookmark `db`:

```python
In [2]: cd db
(bookmark:db) -> /home/wesm/Dropbox/
/home/wesm/Dropbox
```

This automatically takes you to the directory associated with the bookmark.

**Handling Conflicts:**

In case a bookmark name conflicts with a directory name in your current working directory, you can use the `-b` flag to explicitly use the bookmark location.

**Listing Bookmarks:**

To view a list of all your bookmarks, you can use the `-l` option with the `%bookmark` command:

```python
In [3]: %bookmark -l
Current bookmarks:
db -> /home/wesm/Dropbox/
```

**Persistence:**

Unlike aliases, bookmarks are automatically persisted between IPython sessions. This means you won't lose your defined bookmarks when you exit IPython and reopen it later.

IPython's directory bookmarking system streamlines directory navigation and is especially useful for users who frequently work across different directories.


# Timing Code: %time and %timeit


**Summary: Measuring Code Execution Time**

When working on larger-scale or longer-running data analysis tasks, it's often important to measure the execution time of different components, statements, or function calls. IPython simplifies this process by providing two magic functions: `%time` and `%timeit`.

| Magic Function | Purpose |
|----------------|---------|
| `%time` | Measures and reports the total execution time of a statement once. |
| `%timeit` | Runs a statement multiple times to produce an average runtime, providing a more accurate measurement. |

**Example Usage:**

Suppose we want to compare the performance of two methods for selecting strings starting with a specific prefix from a large list of strings:

```python
strings = ['foo', 'foobar', 'baz', 'qux', 'python', 'Guido Van Rossum'] * 100000

# Method 1
method1 = [x for x in strings if x.startswith('foo')]

# Method 2
method2 = [x for x in strings if x[:3] == 'foo']
```

By using `%time` and `%timeit`, we can quickly gauge the execution times of these methods:

```python
%time method1 = [x for x in strings if x.startswith('foo')]  # Wall time: 0.19 s
%time method2 = [x for x in strings if x[:3] == 'foo']      # Wall time: 0.09 s

%timeit [x for x in strings if x.startswith('foo')]        # Average: 159 ms per loop
%timeit [x for x in strings if x[:3] == 'foo']            # Average: 59.3 ms per loop
```

Using `%timeit` is particularly useful for analyzing short-duration statements, even down to the microsecond or nanosecond level. These seemingly insignificant time differences can accumulate, impacting overall performance.

Understanding the performance characteristics of libraries like the Python standard library, NumPy, and pandas is essential, as even small variations in execution times can have a substantial impact in larger-scale data analysis applications.

---

In summary, IPython's `%time` and `%timeit` magic functions provide a convenient way to measure and compare code execution times. By analyzing performance characteristics, you can optimize your code for more efficient data analysis.

# Basic Profiling: %prun and %run -p
**Profiling Code Execution**

Profiling code involves analyzing where execution time is spent in order to identify performance bottlenecks. IPython provides tools for profiling code, such as the cProfile module and the line_profiler extension, which help determine time distribution among functions and even individual lines.

- **cProfile:** The cProfile module is a standard Python profiling tool. It tracks the time spent in each function while executing a program or code block. It can be run from the command line with options to sort and display results. Example:

  ```bash
  python -m cProfile -s cumulative script.py
  ```

  It's useful for macro-level profiling to identify functions consuming the most time.

- **%prun:** In IPython, `%prun` is used to profile code statements or function calls. It provides similar functionality to cProfile but within an IPython session. Example:

  ```python
  %prun -l 7 -s cumulative run_experiment()
  ```

- **line_profiler:** The line_profiler extension allows line-by-line profiling of functions. It's especially useful for micro-level profiling to analyze individual lines of code. To use it:

  1. Install the `line_profiler` package.
  2. Activate the extension by adding `'line_profiler'` to IPython configuration.
  3. Use `%lprun -f func_name` to profile specific functions.

- **Profiling Strategy:** As a guideline, use `%prun` (cProfile) for macro-level profiling to identify functions consuming more time and `%lprun` (line_profiler) for micro-level profiling, understanding performance at the line level. Specifying functions explicitly in `%lprun` helps reduce the overhead of tracing each line.

By using these profiling tools, you can gain insights into code performance and optimize it for efficient data analysis.



# Profiling a Function Line-by-Line
**Summary: Line-by-Line Profiling with `%lprun`**

Profiling code execution is crucial for identifying performance bottlenecks and optimizing software. While traditional profiling methods like `%prun` and cProfile provide insights into function-level performance, complex code or the need for a more detailed analysis may require a different approach. This is where the `line_profiler` library comes into play.

**Enabling the Extension:**
To harness the power of `%lprun`, install the `line_profiler` package and activate the IPython extension by modifying your IPython configuration:

```python
c.TerminalIPythonApp.extensions = ['line_profiler']
```

**How to Use:**
1. Define the functions you want to analyze within your module or IPython session.
2. Employ `%lprun -f function_name statement_to_profile` to profile specific functions.

**Practical Example:**
Imagine a module named `prof_mod` containing functions `add_and_sum` and `call_function` that operate on NumPy arrays:

1. Use `%run` to load the module and set variables.
2. Profile the `add_and_sum` function using `%lprun`:

```python
%run prof_mod
x = randn(3000, 3000)
y = randn(3000, 3000)
%lprun -f add_and_sum add_and_sum(x, y)
```

3. `%lprun` reveals a detailed line-by-line breakdown of execution time for the specified function:

```python
Timer unit: 1e-06 s
File: prof_mod.py
Function: add_and_sum at line 3
Total time: 0.045936 s
Line # Hits Time Per Hit % Time Line Contents
==============================================================
3 def add_and_sum(x, y):
4 1 36510 36510.0 79.5 added = x + y
5 1 9425 9425.0 20.5 summed = added.sum(axis=1)
6 1 1 1.0 0.0 return summed
```

**Choosing Profiling Tools:**
- **Macro-Level Profiling:** Utilize `%prun` (cProfile) to gain an overview of function execution time and identify major performance hotspots.
- **Micro-Level Profiling:** Opt for `%lprun` (line_profiler) for detailed line-by-line analysis, especially when examining functions with very short execution times.

By having proficiency in both macro and micro profiling tools, developers can effectively optimize code and enhance overall software performance.
