<table>
  <tr>
    <th>#</th>
    <th>Measure</th>
    <th>Description</th>
    <th>When/Why to Use</th>
    <th>Formula</th>
  </tr>
  <tr>
    <td>1</td>
    <td>Mean</td>
    <td>The average of all values in a dataset.</td>
    <td>Preferred when the data is normally distributed and does not have extreme outliers. Sensitive to extreme values.</td>
    <td>Mean = Σxi / n</td>
  </tr>
  <tr>
    <td>2</td>
    <td>Median</td>
    <td>The middle value of a dataset when it is ordered from least to greatest.</td>
    <td>Preferred when the data contains outliers or is not normally distributed. Less sensitive to extreme values.</td>
    <td>No specific formula (calculated by sorting and finding the middle value)</td>
  </tr>
  <tr>
    <td>3</td>
    <td>Mode</td>
    <td>The value that appears most frequently in a dataset.</td>
    <td>Suitable for categorical data and discrete distributions. Useful for identifying the most common category or value.</td>
    <td>No specific formula (identified by counting occurrences)</td>
  </tr>
  <tr>
    <td>4</td>
    <td>Midrange</td>
    <td>The average of the maximum and minimum values in a dataset.</td>
    <td>Simple measure useful for small datasets. Can be affected by extreme values.</td>
    <td>Midrange = (Max + Min) / 2</td>
  </tr>
  <tr>
    <td>5</td>
    <td>Geometric Mean</td>
    <td>The nth root of the product of n values in a dataset.</td>
    <td>Appropriate for data with multiplicative relationships, such as financial data or growth rates. Reduces the impact of very large or very small values.</td>
    <td>Geometric Mean = √(x1 * x2 * ... * xn)</td>
  </tr>
  <tr>
    <td>6</td>
    <td>Harmonic Mean</td>
    <td>The reciprocal of the arithmetic mean of the reciprocals of a set of values.</td>
    <td>Useful for averaging rates or ratios, like speed or efficiency. Sensitive to small values, making it less influenced by outliers.</td>
    <td>Harmonic Mean = n / (1/x1 + 1/x2 + ... + 1/xn)</td>
  </tr>
  <tr>
    <td>7</td>
    <td>Weighted Mean</td>
    <td>A mean where different values have different weights or importance.</td>
    <td>Useful when certain data points contribute more significance to the overall average. Allows customization of the influence of each data point.</td>
    <td>Weighted Mean = (Σwi * xi) / Σwi</td>
  </tr>
  <tr>
    <td>8</td>
    <td>Trimean</td>
    <td>The average of the median and the midrange.</td>
    <td>Combines characteristics of the median and midrange, providing a balance between resistance to outliers and capturing central tendency.</td>
    <td>Trimean = (Median + Midrange) / 2</td>
  </tr>
  <tr>
    <td>9</td>
    <td>Winsorized Mean</td>
    <td>A modified mean that limits the impact of outliers by replacing extreme values with values closer to the mean.</td>
    <td>Useful when outliers have a significant impact on the standard mean. Reduces the influence of extreme values without entirely excluding them.</td>
    <td>Simple description: Replaces extreme values with values closer to the mean</td>
  </tr>
  <tr>
    <td>10</td>
    <td>MAD (Mean Absolute Deviation)</td>
    <td>The average of the absolute differences between each data point and the mean.</td>
    <td>Offers insight into the average distance between data points and the mean. Robust to outliers, providing a measure of dispersion.</td>
    <td>MAD = Σ|xi - Mean| / n</td>
  </tr>
  <tr>
    <td>11</td>
    <td>L-Trimmed Mean</td>
    <td>A mean calculated after removing a certain percentage of the lowest and highest values.</td>
    <td>Provides a compromise between the mean and the trimmed mean. Reduces the impact of extreme values while maintaining some sensitivity to central tendency.</td>
    <td>Simple description: Removes a certain percentage of extreme values</td>
  </tr>
  <tr>
    <td>12</td>
    <td>Huber Loss</td>
    <td>A measure used in robust regression that combines features of the mean and the median.</td>
    <td>Balances the robustness of the median with the efficiency of the mean in regression analysis. Useful when there are potential outliers in the data.</td>
    <td>Simple description: Robust regression measure</td>
  </tr>
  <tr>
    <td>13</td>
    <td>Quadratic Mean (Root Mean Square)</td>
    <td>The square root of the average of the squared values in a dataset.</td>
    <td>Commonly used in signal processing and physics to calculate the magnitude of a set of values. Reflects the dispersion of values and emphasizes larger values.</td>
    <td>Quadratic Mean = √((x1^2 + x2^2 + ... + xn^2) / n)</td>
  </tr>
  <tr>
    <td>14</td>
    <td>Bi-weight Midvariance</td>
    <td>A measure of statistical dispersion that combines features of the median and variance, giving less weight to outliers.</td>
    <td>Useful for robust estimation of variance, balancing resistance to outliers with sensitivity to the spread of data.</td>
    <td>Simple description: Robust variance estimator</td>
  </tr>
  <tr>
    <td>15</td>
    <td>Gini Mean Difference</td>
    <td>A measure of statistical dispersion based on Gini index, used to assess inequality in a dataset.</td>
    <td>Applicable when assessing economic inequality or distribution of resources. A higher Gini Mean Difference indicates greater inequality.</td>
    <td>Gini Mean Difference = (1/n^2) Σ Σ |xi - xj|</td>
  </tr>
  <tr>
    <td>16</td>
    <td>Variance</td>
    <td>The average of the squared differences from the mean in a dataset.</td>
    <td>Provides a measure of data spread. Suitable for datasets without extreme outliers. Sensitive to squared deviations, which can amplify the impact of extreme values.</td>
    <td>Variance = Σ(xi - Mean)^2 / n</td>
  </tr>
  <tr>
    <td>17</td>
    <td>Coefficient of Variation (CV)</td>
    <td>The ratio of the standard deviation to the mean, often expressed as a percentage, providing a relative measure of variability.</td>
    <td>Useful for comparing the relative variability of datasets with different units or scales. A lower CV indicates more consistent data.</td>
    <td>CV = (SD / Mean) * 100%</td>
  </tr>
  <tr>
    <td>18</td>
    <td>Trimmed Mean</td>
    <td>Similar to L-Trimmed Mean but involves removing a certain percentage of data from both ends, not just one.</td>
    <td>Offers a compromise between the mean and median. Reduces the influence of extreme values while maintaining some sensitivity to central tendency.</td>
    <td>Simple description: Removes a certain percentage of extreme values from both ends</td>
  </tr>
  <tr>
    <td>19</td>
    <td>Qn (Sn Scale)</td>
    <td>A robust measure of scale (dispersion) similar to the interquartile range but with a more efficient breakdown point.</td>
    <td>Robust to outliers and resistant to skewed data. Suitable for datasets with potential extreme values.</td>
    <td>Qn = c * (Q3 - Q1) / 1.346</td>
  </tr>
  <tr>
    <td>20</td>
    <td>Harrell-Davis Quantile Estimator</td>
    <td>A non-parametric estimator of the quantiles of a population.</td>
    <td>Useful for estimating quantiles in datasets with unknown or complex distributions, especially when other measures might be influenced by outliers.</td>
    <td>Simple description: Non-parametric quantile estimator</td>
  </tr>
</table>
