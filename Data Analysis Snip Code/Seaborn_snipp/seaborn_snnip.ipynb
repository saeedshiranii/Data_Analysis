## get names of seaborn datasets

```python
import seaborn as sns
```

```python
# this is name of datasets which are provided by seaborn
sns.get_dataset_names()
```

## Load datasets

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips.head()
```

```python
tips.columns
```

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time', # facid grid
    hue='smoker',
    style='smoker',
    size='size'
)
    
```

```python
dots = sns.load_dataset("dots")
dots.columns
```

```python
sns.relplot(
    data=dots,
    kind='line',
    x='time', 
    y='firing_rate',
    col='align',
    hue='choice',
    size='coherence',
    style='choice',
    facet_kws=dict(sharex=False) # when scale of data on two axis is not equal, then we need to use facet_kws
    
    
)
```

# linear estimation with seaborn

```python
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    hue='smoker'
)
```

# Informative Disturbayional summeries

```python
sns.displot(
    data=tips,
    x='total_bill',
    col='time',
    kde=True # this draw a curv
)
```

```python
sns.displot(
    data=tips,
    x='total_bill',
    col='time',
    hue='smoker',
    kind='ecdf',
    rug=True # this draw a curv
);
# this will show you time of each event
```

# categorical data

```python
sns.catplot(
    data=tips,
    kind='swarm',
    x='day',
    y='total_bill',
    hue='smoker',
    dodge=True
);
```

```python
sns.catplot(
    data=tips,
    kind="violin",
    x="day",
    y="total_bill",
    hue="smoker",
    split=True
    
);
```

```python
sns.catplot(
    data=tips,
    kind='bar',
    x='day',
    y='total_bill',
    hue='smoker'
);
```
