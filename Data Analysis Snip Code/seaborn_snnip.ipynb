## Load datasets

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips.head()
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue='day', size='size', palette="YlGnBu")
```

```python
sns.histplot(tips['tip'], kde=True, bins=20)
```

```python
sns.barplot(x= 'sex', y = 'total_bill', data=tips, palette='YlGnBu')
```

```python
sns.boxplot(x="day", y="tip", data=tips, palette="YlGnBu", hue="sex") 
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="YlGnBu", hue="sex") 
```
