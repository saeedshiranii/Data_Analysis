<table>

    <tr>
      <th>Wildcard</th>
      <th>Description</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td>%</td>
      <td>Represents zero or more characters. For example, <code>LIKE 'a%'</code> finds values starting with "a". <code>LIKE '%or%'</code> finds values with "or" in any position.</td>
    </tr>
    <tr>
      <td>_</td>
      <td>Represents a single character. For example, <code>LIKE '_r%'</code> finds values with "r" in the second position. <code>LIKE 'a_%_%'</code> matches values starting with "a" and at least 3 characters long.</td>
    </tr>
    <tr>
      <td>[charlist]</td>
      <td>Represents any single character in the specified list. For instance, <code>LIKE '[a-c]%'</code> matches values starting with "a", "b", or "c".</td>
    </tr>
    <tr>
      <td>[^charlist]</td>
      <td>Represents any single character not in the specified list. For example, <code>LIKE '[^a-c]%'</code> matches values not starting with "a", "b", or "c".</td>
    </tr>
    <tr>
      <td>[char1-char2]</td>
      <td>Represents any single character in the range between char1 and char2. E.g., <code>LIKE '[0-9]%'</code> matches values starting with a digit.</td>
    </tr>
    <tr>
      <td>[^char1-char2]</td>
      <td>Represents any single character not in the range between char1 and char2. For instance, <code>LIKE '[^0-9]%'</code> matches values not starting with a digit.</td>
    </tr>
  </tbody>

Certainly! Below is a basic cheat sheet for working with stored procedures in MySQL, specifically for use in a Jupyter Notebook environment. Make sure to run these cells in a Jupyter Notebook with a MySQL connection.

### Creating a Stored Procedure:

```sql
DELIMITER //

CREATE PROCEDURE procedure_name (IN parameter1 datatype1, IN parameter2 datatype2, OUT output_parameter datatype)
BEGIN
    -- SQL statements to define the procedure logic
END //

DELIMITER ;
```

Replace `procedure_name`, `parameter1`, `datatype1`, `parameter2`, `datatype2`, `output_parameter`, and `datatype` with your actual values.

### Example:

```sql
DELIMITER //

CREATE PROCEDURE GetEmployeeCount(IN department_id INT, OUT employee_count INT)
BEGIN
    SELECT COUNT(*) INTO employee_count FROM employees WHERE department_id = department_id;
END //

DELIMITER ;
```

### Calling a Stored Procedure:

```sql
CALL procedure_name(argument1, argument2, ...);
```

Replace `procedure_name` and `argument1`, `argument2`, etc., with your actual values.

### Example:

```sql
CALL GetEmployeeCount(1, @count);
SELECT @count AS EmployeeCount;
```

### Altering a Stored Procedure:

```sql
DELIMITER //

ALTER PROCEDURE procedure_name
MODIFY parameter1 datatype1,
MODIFY parameter2 datatype2
BEGIN
    -- Updated SQL statements
END //

DELIMITER ;
```

### Example:

```sql
DELIMITER //

ALTER PROCEDURE GetEmployeeCount
MODIFY department_id INT,
MODIFY employee_count INT
BEGIN
    -- Updated SQL statements
END //

DELIMITER ;
```

### Dropping a Stored Procedure:

```sql
DROP PROCEDURE IF EXISTS procedure_name;
```

Replace `procedure_name` with the actual name of your stored procedure.

### Example:

```sql
DROP PROCEDURE IF EXISTS GetEmployeeCount;
```

Feel free to customize these examples according to your specific requirements. This cheat sheet covers basic syntax for creating, calling, altering, and dropping stored procedures in MySQL.
